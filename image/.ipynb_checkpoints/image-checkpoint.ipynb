```python
# 시작할때 이거부터 쓰기
import cv2 #영상 처리 라이브러리
import matplotlib.pyplot as plt #시각화(이미지 띄우기 등)
import numpy as np # 배열 처리
```

```python
# 실습 3 논리연산
# 이미지 읽어오기
img = cv2.imread("images/Lenna.png",cv2.IMREAD_GRAYSCALE)
h,w = img.shape

# 원모양의 마스크생성하기
mask = cv2.circle(np.zeros((h,w)), (np.uint8(h/2), np.uint8(w/2)), 80, 255, -1) 
# 검정색 네모, 원의 중심 값, 원의 반지름, color (흰색)
# 나눌 때 cv2 는 정수형만 취급하기 때문에 int(), uint8 형으로 감싸줘야 함.
mask = np.uint8(mask)

plt.imshow(mask, cmap="gray")
# cv2 사용하여 논리연산 수행하기
and_img = cv2.bitwise_and(img,np.uint8(mask)) #and 연산
or_img = cv2.bitwise_or(img,np.uint8(mask)) #or 연산
xor_img = cv2.bitwise_xor(img,np.uint8(mask)) #xor 연산
not_img = cv2.bitwise_not(img,np.uint8(mask)) #not 연산
# 시각화
plt.subplot(2,3,1); plt.imshow(img, cmap="gray")
plt.subplot(2,3,2); plt.imshow(and_img, cmap="gray")
plt.subplot(2,3,3); plt.imshow(or_img, cmap="gray")
plt.subplot(2,3,4); plt.imshow(xor_img, cmap="gray")
plt.subplot(2,3,5); plt.imshow(not_img, cmap="gray")
```

```python
# 실습4

# null transform : y = 1*x +0
null_img = np.zeros((h,w))
a=1
b=0
for x in range(0,h):
    for y in range(0,w):
        null_img[x,y] = a * img[x,y] + b
plt.subplot(1,2,1); plt.imshow(img, cmap="gray")
plt.subplot(1,2,2); plt.imshow(null_img, cmap="gray")




```

```python
# negative transform : y= -1*x + 255
nega_img = np.zeros((h,w))
a=-1
b=255
for x in range(0,h):
    for y in range(0,w):
        nega_img[x,y] = a * img[x,y] + b
plt.subplot(1,2,1); plt.imshow(img, cmap="gray")
plt.subplot(1,2,2); plt.imshow(nega_img, cmap="gray")
```

```python
# gamma correction : y = (x/255)r* 255(r=4)

a=-1
b=255
gamma = 4
gamma_img = np.zeros((h,w))
for x in range(0,h):
    for y in range(0,w):
        gamma_img[x,y] = (img[x,y]/255)**(gamma) * 255
        
gamma_img =np.uint8(gamma_img) 
plt.subplot(1,2,1); plt.imshow(img, cmap="gray")
plt.subplot(1,2,2); plt.imshow(gamma_img, cmap="gray")
```

```python
# intensity contrast stretch  y = a·x + b (a, b = ?)
# x축 범위 : 50~200, y축 범위 : 0~255(단, y가 0보다 작으면 0, 255보다 크면 255)
ics_img = np.zeros((h,w))
a = (255-0)/(200-50) # 기울기
b = -a*50 # y절편
print(a,b)
for x in range(0,h):
    for y in range(0,w):
        if img[x,y] < 50:
            ics_img[x,y] = 0
        elif img[x,y] > 200:
            ics_img[x,y] = 255
        else:
            ics_img[x,y] = a * img[x,y] + b
plt.subplot(1,2,1); plt.imshow(img, cmap="gray")
plt.subplot(1,2,2); plt.imshow(ics_img, cmap="gray")
```

```python
# intensity contrast compress : y = a·x + b (a, b = ?)
#     x축 범위 : 0~255, y축 범위 : 50~200

icc_img = np.zeros((h,w))
a = (200-50)/(255-0)
b = a *50
print(a,b)
for x in range(0,h):
    for y in range(0,w):
        if img[x,y] < 50:
            icc_img[x,y] = 255
        elif img[x,y] > 200:
            icc_img[x,y] = 0
        else:
            icc_img[x,y] = a * img[x,y] + b  
plt.subplot(1,2,1); plt.imshow(img, cmap="gray")
plt.subplot(1,2,2); plt.imshow(icc_img, cmap="gray")
```
